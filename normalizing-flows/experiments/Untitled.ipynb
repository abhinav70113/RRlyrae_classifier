```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
INPUT = 200000
INTEREST_PERCENT = 9
TIME = 30
```

```python
time = np.arange(1, TIME+1)

fig = plt.figure(figsize=(21, 29.7))
plt.subplots_adjust(hspace=0.15, wspace=0.1)

for i, INPUT in enumerate([60, 80, 100, 120, 150, 180, 240, 300]):
    INPUT *= 1000
    
    ax = fig.add_subplot(4, 2, i+1)
    
    for INTEREST_PERCENT in range(5,15):
        interest = 1 + INTEREST_PERCENT / 100

        value = INPUT * interest**time
        
        LINE_WIDTH = np.abs(4.5 - np.abs(9-INTEREST_PERCENT)) / 2

        ax.plot(time, value*1e-5, label=f"{INTEREST_PERCENT} %", lw=LINE_WIDTH, color="r")

    ax.legend(loc="upper left", frameon=False, shadow=True, title="interest:")
    ax.grid(alpha=.5, ls=":")

    ax.set_title(f"initial capital: {int(INPUT/1000)} 000 ", fontweight=55)
    ax.set_xlabel("time after investment")
    ax.set_ylabel("value / 100 000 ")

    ax.set_xticks(time) # steps of `1`
    max_val = int((value*1e-5).max()) + 3
    steps = 1
    if max_val > 30:
        steps = 2
    ax.set_yticks(np.arange(0, max_val, steps)) # steps of `1`

plt.show()
```
